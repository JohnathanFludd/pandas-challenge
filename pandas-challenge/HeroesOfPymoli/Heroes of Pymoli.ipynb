```python
# Dependencies and Setup
import pandas as pd
import numpy as np
# File to Load (Remember to Change These)
file = "../purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file)
df.head()
```

```python
#Total Number Of Players

total_players = df["Purchase ID"].count()

#print("Total Number Of Players: " + str(total_players))


#Purchase Count

purchase_count = df["Item ID"].count()

#Number Of Unique Items

df_unique_items = df["Item ID"].unique()
df_unique_items = len(df_unique_items)
#print("Number Of Unique Items: " + str(df_unique_items))

#Average Purchase Price

avg_price = df["Price"].mean()
avg_price = ("${:.2f}".format(avg_price)) 
#print("Average Purchase Price : " + str(avg_price))

#Total Revenue

total_rev = df["Price"].sum()
total_rev = ("${:.2f}".format(total_rev))
#print("Total Revenue : " + str(total_rev))


#Purchasing Analysis (Total)

Purchasing_Analysis = {
     "Number Of Unique Items": [df_unique_items],
     "Average Purchase Price": [avg_price],
     "Total Number Of Purchases": [purchase_count],
     "Total Revenue": [total_rev]
 }
Purchasing_Analysis_df = pd.DataFrame(Purchasing_Analysis)
Purchasing_Analysis_df
```

```python
#Gender Demographics

mf_df = df['Gender']
unique_mf_df = df[['SN', 'Gender']].drop_duplicates(subset ="SN", keep = 'first', inplace = False)
df_drop = df.drop_duplicates(subset ="SN", keep = 'first', inplace = False)


#Total Unique Players

total_unique_players = total_players = df_drop["SN"].count()


unique_total_mf_df = unique_mf_df["Gender"].value_counts()


unique_total_mf_df = unique_mf_df["Gender"].value_counts()


unique_total_mf_df = pd.DataFrame(unique_total_mf_df)

unique_total_mf_df['Percentage of Players'] = (unique_total_mf_df['Gender'] / sum(unique_total_mf_df['Gender'])).map("{:,.2%}".format)
unique_total_mf_df.rename(columns={"Gender": "Total Count"}, inplace = True)
unique_total_mf_df
```

```python
#Purchasing Analysis (Male)

df_m = df[df["Gender"] == "Male"]

total_price_m = df_m["Price"].sum()


total_price_m = df_m["Price"].sum()
total_count_m = df_m.shape[0]

avg_price_m = (total_price_m /total_count_m)
avg_price_m = ("${:.2f}".format(avg_price_m))


total_m_avg_purchase = df_m.groupby("SN")["Price"].sum().mean()
total_m_avg_purchase = ("${:.2f}".format(total_m_avg_purchase))


result_df_m = pd.DataFrame({"Gender": ["Male"],
                          "Purchase Count": [total_count_m],
                          "Average Purchase Price": [avg_price_m],
                          "Total Purchase Value": [total_price_m],
                          "Avg Total Purchase per Person": [total_m_avg_purchase]
                         })


total_mf = pd.DataFrame(result_df_m)

total_mf.rename(columns={"Gender": "Total Count"}, inplace = True)

# result_df_m

#---------------------------------------------------------------------------------------------------------------------------
#Purchasing Analysis (Female)

df_f = df[df["Gender"] == "Female"]

total_price_f = df_f["Price"].sum()
total_count_f = df_f.shape[0]


avg_price_f = (total_price_f /total_count_f)
avg_price_f = ("${:.2f}".format(avg_price_f))


total_f_avg_purchase = df_f.groupby("SN")["Price"].sum().mean()
total_f_avg_purchase = ("${:.2f}".format(total_f_avg_purchase))


result_df_f = pd.DataFrame({"Gender": ["Female"],
                          "Purchase Count": [total_count_f],
                          "Average Purchase Price": [avg_price_f],
                          "Total Purchase Value": [total_price_f],
                          "Avg Total Purchase per Person": [total_f_avg_purchase]
                         })


total_mf = pd.DataFrame(result_df_f)

total_mf.rename(columns={"Gender": "Total Count"}, inplace = True)

#result_df_f

#---------------------------------------------------------------------------------------------------------------------------
#Purchasing Analysis (Other)

df_o = df[df["Gender"] == "Other / Non-Disclosed"]

total_price_o = df_o["Price"].sum()
total_count_o = df_o.shape[0]


avg_price_o = (total_price_o /total_count_o)
avg_price_o = ("${:.2f}".format(avg_price_o))


total_o_avg_purchase = df_o.groupby("SN")["Price"].sum().mean()
total_o_avg_purchase = ("${:.2f}".format(total_o_avg_purchase))


result_df_o = pd.DataFrame({"Gender": ["Other / Non-Disclosed"],
                          "Purchase Count": [total_count_o],
                          "Average Purchase Price": [avg_price_o],
                          "Total Purchase Value": [total_price_o],
                          "Avg Total Purchase per Person": [total_o_avg_purchase]
                         })


total_mf = pd.DataFrame(result_df_o)

total_mf.rename(columns={"Gender": "Total Count"}, inplace = True)

#result_df_o
```

```python
#Combine Purchase Analysis Gender Charts

pd.concat([result_df_f, result_df_m, result_df_o],ignore_index = True, sort = False).set_index("Gender", drop = True)
```

```python
#Age Demographics


# Create the bins in which Data will be held
  
bins = [4.9, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 44.9]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


df[""] = pd.cut(df["Age"], bins, labels=group_names, include_lowest=True)
#df
df_drop["Age Demographics"] = pd.cut(df_drop["Age"], bins, labels=group_names, include_lowest=True)
df_drop
```

```python
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table

total_ages = df_drop.groupby("Age Demographics")["SN"].count()

result_df_ages = pd.DataFrame(total_ages)
result_df_ages["Age Percentages"] = (result_df_ages["SN"] / total_unique_players).map("{:,.2%}".format)
result_df_ages.rename(columns={"SN": "Total Count"}, inplace = True)
result_df_ages.rename(columns={"Age Percentages": "Percentage of Players"}, inplace = True)
result_df_ages.index.name = None


#total_ages
result_df_ages
```

```python
#Purchasing Analysis (Age)

age_df = df_drop.groupby("Age Demographics")["Price"].agg(PurchaseCount="count",AvgPurchasePrice="mean", TotalPurchaseValue="sum")
age_df["Avg Total Purchase per Person"] = (age_df["TotalPurchaseValue"] / total_ages).map("${:.2f}".format)

age_df["AvgPurchasePrice"] = age_df["AvgPurchasePrice"].map("${:.2f}".format)
age_df["PurchaseCount"] = age_df["PurchaseCount"].map("${:.2f}".format)
age_df["TotalPurchaseValue"] = age_df["TotalPurchaseValue"].map("${:.2f}".format)
#age_df[""]

#total_ages
result_df_ages_prices = pd.DataFrame(total_ages)
result_df_ages_prices.index.name = None
#result_df_ages_prices
age_df
```

```python
# Top Spenders

# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame



top_age_df = df.groupby(["SN"])["Price"].agg(PurchaseCount="count",AvgPurchasePrice="mean", TotalPurchaseValue="sum")
top_age_df = top_age_df.sort_values(by=['TotalPurchaseValue'], ascending=False)

top_age_df["AvgPurchasePrice"] = top_age_df["AvgPurchasePrice"].map("${:.2f}".format)
top_age_df["TotalPurchaseValue"] = top_age_df["TotalPurchaseValue"].map("${:.2f}".format)
top_age_df.head()
```

```python
# Most Popular Items
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame



pop_items_df = df.groupby(["Item ID", "Item Name"])["Price"].agg(PurchaseCount="count",AvgPurchasePrice="mean", TotalPurchaseValue="sum")
pop_items_df = pop_items_df.sort_values(by=['PurchaseCount'], ascending=False)

pop_items_df["AvgPurchasePrice"] = pop_items_df["AvgPurchasePrice"].map("${:.2f}".format)
pop_items_df["TotalPurchaseValue"] = pop_items_df["TotalPurchaseValue"].map("${:.2f}".format)

pop_items_df.head()
```

```python
# Most Profitable Items
# Sort the above table by total purchase value in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the data frame

 
top_items_df = df.groupby(["Item ID", "Item Name"])["Price"].agg(PurchaseCount="count",AvgPurchasePrice="mean", TotalPurchaseValue="sum")


top_items_df["AvgPurchasePrice"] = top_items_df["AvgPurchasePrice"].map("${:.2f}".format)
top_items_df["TotalPurchaseValue"] = top_items_df["TotalPurchaseValue"].map("${:.2f}".format)

top_items_df = pop_items_df.sort_values(by=['TotalPurchaseValue'], ascending=False)

top_items_df.head()
```
